```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df_1= pd.read_csv('s3.csv')
df_1.head()
```

```python
x_factor =df_1.iloc[:,:-1].values
x_factor
x_factor.shape
```

```python
y_factor=df_1.iloc[:,-1].values
```

```python
imputerObj= SimpleImputer(missing_values =np.nan, strategy='median')
```

```python
imputerObj.fit(x_factor[:,1:3])
x_factor[:,1:3]=imputerObj.fit_transform(x_factor[:,1:3])
x_factor
```

```python

x_factor = df_1.iloc[:, :-1].values
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop = 'first'), [0])], remainder='passthrough')
x_factor=np.array(ct.fit_transform(x_factor))
x_factor
```

```python

le = LabelEncoder()
y_factor = le.fit_transform(y_factor)
y_factor
```

```python
train_test_split(x_factor,y_factor,test_size=0.2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_factor, y_factor, test_size = 0.2)
X_train, X_test, y_train, y_test
```
